```python
modelFolder = './'
import os
from numpy import genfromtxt
import numpy as np
import csv
ipath = modelFolder + 'stims'
opath = 'stims_plus400'
input_list = os.listdir(ipath)
```

```python
def concatenate(pad):
    zeros = np.zeros(pad)
    for stim in input_list:
        if stim.endswith('.csv') == False:
            continue
        stim_fn = ipath + '/' + stim
        print(stim_fn)
        curr_stim = genfromtxt(stim_fn, delimiter=',')
        padded_stim = np.concatenate((curr_stim, zeros), axis=0)
        print(padded_stim)
        with open(opath + '/' + stim, 'w') as f:
            writer = csv.writer(f)
            writer.writerows(map(lambda x: [x], padded_stim))
```

```python
concatenate(400)
```
